```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

```python
# minimum and maximum timeseries length
MIN = 700
MAX = 1200

# generating random walks
num_RW = 50 # number of random walks
RW = [] # list of all random walks

for _ in range(num_RW):
    
    size = np.random.randint(MIN,MAX)
    rw = np.zeros(size)
    rw[0] = np.random.normal(0,1)
    
    for i in range(1,size):
        rw[i] = rw[i-1] + np.random.normal(0,1)
        
    RW.append(pd.Series(rw))
    
    
# generating sin waves
num_SIN = 40 # number of sin waves
SIN = []

for _ in range(num_SIN):
    
    size = np.random.randint(MIN, MAX)
    x = np.linspace(0,500, size)
    f = np.random.randint(25,75)
    sw = np.random.normal(0,1)*np.sin(x/f)
    SIN.append(pd.Series(sw))
```

```python
plt.figure(0)
for i in RW:
    i.plot()

plt.figure(1)
for j in SIN:
    j.plot()
```
